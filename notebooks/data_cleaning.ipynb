# Preprocessing Features

## Preprocessing All features with [Nan, standart equipment, optinal equipment]
saved in: columns_to_transform ---------
Nan = 0 ----------------
standart equipment, optinal equipment = 1

```python
preprocessing_df = new_filtered_df.copy()

columns_to_transform = ['Hot air', 'Rev counter', 'Assisted direction', 'ABS brakes', 'Rear window', 'Central locking of the doors',
                        'Headrest for all occupants', 'Electric rearview mirror adjustment', 'Air conditioning',
                        'Bluetooth connection', 'Frontal Airbags', 'Steering wheel adjustment height',
                        'Electric front window control', 'Multifunctional steering wheel', 'Driver s seat with height adjustment',
                        'On board computer', 'Light in the trunk', 'Alloy wheels', 'USB connection', 'Radio',
                        'Folding rear seat', 'Perimeter anti theft alarm', 'Cooling liquid thermometer']

# Replace NaN with 0 and other values with 1
preprocessing_df[columns_to_transform] = preprocessing_df[columns_to_transform].applymap(lambda x: 0 if pd.isnull(x) else 1)
```

## Preprocessing [Acceleration]

```python
# Preprocess Acceleration (0100 km/h 3,8 s = 3.8)
preprocessing_df['Acceleration'] = preprocessing_df['Acceleration'].str.extract(r'(\d+\,\d+)')

# Replace ',' with '.' and convert to numeric
preprocessing_df['Acceleration'] = preprocessing_df['Acceleration'].str.replace(',', '.').astype(float)
# Rename Column
preprocessing_df = preprocessing_df.rename(columns={'Acceleration': 'Acceleration 0100 km/h in S'})
```

```python
# Clean 'Car payload'
preprocessing_df['Car payload'] = preprocessing_df['Car payload'].str.replace(' kg', '', regex=True).astype('Int64')

# Clean 'Specific power'
preprocessing_df['Specific power'] = preprocessing_df['Specific power'].str.replace(' cv/litro', '', regex=True)
preprocessing_df['Specific power'] = preprocessing_df['Specific power'].str.replace(',', '.', regex=True).astype(float)

# Clean 'CNW Index'
# preprocessing_df['CNW Index'] = preprocessing_df['CNW Index'].str.replace(',', '.', regex=True)

# Clean 'Reader score'
preprocessing_df['Reader score'] = preprocessing_df['Reader score'].str.replace(',', '.', regex=True).astype(float)

# Clean 'Sidewall height'
preprocessing_df['Sidewall height'] = preprocessing_df['Sidewall height'].str.replace(' mm', '', regex=True).astype(float)

# Clean 'Provenance' (removing special characters and leading/trailing spaces)
preprocessing_df['Provenance'] = preprocessing_df['Provenance'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
preprocessing_df['Provenance'] = preprocessing_df['Provenance'].str.strip()

# Clean 'Doors'
preprocessing_df['Doors'] = preprocessing_df['Doors'].astype('Int64')

# Clean 'Maximum power'
preprocessing_df['Maximum power'] = preprocessing_df['Maximum power'].str.replace('[^0-9.]', '', regex=True).astype(float)

# Clean 'Settings'
preprocessing_df['Settings'] = preprocessing_df['Settings'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
preprocessing_df['Settings'] = preprocessing_df['Settings'].str.strip()

# Clean 'Car size'
preprocessing_df['Car size'] = preprocessing_df['Car size'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
preprocessing_df['Car size'] = preprocessing_df['Car size'].str.strip()

# Clean 'Unit displacement'
preprocessing_df['Unit displacement'] = preprocessing_df['Unit displacement'].str.replace('[^0-9.]', '', regex=True).astype(float)

# Clean 'Trunk'
preprocessing_df['Trunk'] = preprocessing_df['Trunk'].str.replace('[a-zA-Z\s]', '', regex=True).astype(float)

# Clean 'Urban'
preprocessing_df['Urban'] = preprocessing_df['Urban'].str.replace(',', '.', regex=True)
preprocessing_df['Urban'] = preprocessing_df['Urban'].str.replace('[^0-9.]', '', regex=True).astype(float)

# Clean 'Year'
preprocessing_df['Year'] = preprocessing_df['Year'].astype('Int64')
```
