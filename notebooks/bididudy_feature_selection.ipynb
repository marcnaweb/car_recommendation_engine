```python
import pandas as pd
import re

def count_columns_with_numbers_and_units(df):
    """
    Counts the number of columns in a DataFrame that contain values with numbers followed by units.

    Parameters:
    - df: pd.DataFrame - The DataFrame to analyze.

    Returns:
    - int - The number of columns containing numbers and units.
    - list - The names of columns containing numbers and units.
    """
    columns_with_units = []
    pattern = re.compile(r'\d+(\.\d+)?\s+[a-zA-Z]+')  # Pattern for numbers followed by units

    for column in df.columns:
        if df[column].dtype == 'object':  # Only check object (string) columns
            if any(df[column].astype(str).apply(lambda x: bool(pattern.search(x)))):
                columns_with_units.append(column)

    return len(columns_with_units), columns_with_units

# Usage example (assuming preprocessing_df is your DataFrame):
# num_columns_with_units, columns_with_units = count_columns_with_numbers_and_units(preprocessing_df)
# print(f"Number of columns with numbers and units: {num_columns_with_units}")
# print(f"Columns with numbers and units: {columns_with_units}")
```

```python
from collections import defaultdict

def find_columns_with_single_unit_dict(df):
    # Regular expression to find units: numbers followed by space (optional) and letters
    pattern = re.compile(r'\d+(\.\d+)?\s*([a-zA-Z]+)')

    # Dictionary to hold unique units found across all columns
    units_found = defaultdict(set)

    # Iterate over each column in the DataFrame
    for column in df.columns:
        if df[column].dtype == 'object':  # Check only string-type columns
            for value in df[column].dropna().unique():
                matches = pattern.findall(value)
                for match in matches:
                    # Extract the unit part
                    if match[1]:  # match[1] contains the unit part from the regex group
                        units_found[column].add(match[1])

    # Create a dictionary for columns that have exactly one unique unit, mapping column names to their unit
    columns_with_single_unit_dict = {column: units.pop() for column, units in units_found.items() if len(units) == 1}

    return columns_with_single_unit_dict
```

```python
def remove_units_and_rename_columns(df, columns_units_dict):
    """
    Removes units from column values based on a provided dictionary mapping columns to their units,
    and renames the columns to include the unit.

    Parameters:
    - df: pd.DataFrame - The DataFrame to process.
    - columns_units_dict: dict - A dictionary where keys are column names and values are units.

    Returns:
    - pd.DataFrame - The DataFrame with units removed from values and columns renamed.
    """
    # Copy the DataFrame to avoid modifying the original
    df_processed = df.copy()

    # Regular expression to find and remove units: numbers followed by optional space and units
    pattern = re.compile(r'(\d+(\.\d+)?)\s*[a-zA-Z]+')

    for column, unit in columns_units_dict.items():
        # Check if the column exists to avoid KeyError
        if column in df_processed.columns:
            # Remove the unit from the column values, keeping only the numeric part
            df_processed[column] = df_processed[column].astype(str).apply(lambda x: pattern.sub(r'\1', x))
            # Convert the cleaned column to numeric type
            df_processed[column] = pd.to_numeric(df_processed[column], errors='coerce')
            # Rename the column to include the unit
            df_processed.rename(columns={column: f"{column} in {unit}"}, inplace=True)

    return df_processed
```

```python
def count_columns_with_rpm(df):

    columns_with_rpm = []

    for column in df.columns:
        if df[column].dtype == 'object':  # Only check object (string) columns
            # Check if any value in the column contains 'kg'
            if df[column].astype(str).str.contains('rpm').any():
                columns_with_rpm.append(column)

    return columns_with_rpm
```

```python
column_list_rpm=count_columns_with_rpm(preprocessing_df)
```

```python
def process_and_rename_columns_in_rpm(df, columns_list):

    # Copy the DataFrame to avoid modifying the original
    df_processed = df.copy()

    for column_name in columns_list:
        # Remove ' rpm' and convert to numeric, handling errors by coercing them to NaN
        df_processed[column_name] = pd.to_numeric(df_processed[column_name].str.replace(' rpm', ''), errors='coerce')

        # Rename the column
        new_name = f"{column_name} in rpm"
        df_processed = df_processed.rename(columns={column_name: new_name})

    return df_processed
```
