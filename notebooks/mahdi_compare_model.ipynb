```python
import pandas as pd

# Load the dataset
file_path = 'unique_car_models_df.csv'
car_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
car_df.head()
```

```python
numeric_features = [
                'car_code',
                'car_model',
                'Car size_Compact and portable',
                'Car size_Coupe',
                'Car size_Grande',
                'Car size_Medium',
                'Car size_Medium Compact',
                'Car size_Subcompact',
                'Length',
                'Width',
                'Weight/Torque',
                'Height',
                'Minimum height from the ground',
                'Weight',
                'Weight/power',
                'Length between the axis',
                'Maximum speed',
                'Steering wheel adjustment height',
                'Reader score',
                'Sidewall height',
                'Trunk',
                'Gear speed transmissions',
                'Urban',
                'Year'
                ]

car_data = car_df[numeric_features]
car_data.head(10)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Exclude 'car_model' column for PCA
features = car_data.drop(columns=['car_model', 'car_code'])

# Initialize PCA, aiming to keep 95% of the variance
pca = PCA(n_components=0.95, random_state=0)

# Fit and transform the feature data
pca_features = pca.fit_transform(features)

# Number of components
n_components = pca.n_components_

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=pca_features, columns=[f'PC{i+1}' for i in range(n_components)])

# Show the number of components and a preview of the PCA DataFrame
n_components, pca_df.head()
```

```python
# Visualization of the first two principal components
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', data=pca_df, s=100, alpha=0.7)
plt.title('PCA of Car Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
# Extracting the PCA components (i.e., loadings)
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(n_components)], index=features.columns)

# Display the loadings of the first few principal components
loadings.head(10)
```

```python
# Select the first 10 principal components for visualization
loadings_subset = loadings.iloc[:, :10]

# Create a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(loadings_subset, cmap="vlag", annot=True, fmt=".2f")
plt.title('Heatmap of Feature Contributions to Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
# Due to the reset before, we need to recompute the most influential features

# Finding the absolute highest loadings for each principal component
highest_loadings = loadings_subset.abs().max().sort_values(ascending=False)

# Identifying the feature(s) that contribute most to each principal component
most_influential_features = loadings_subset.apply(lambda x: x.abs().idxmax())

# Combine the information into a DataFrame for clearer presentation
most_influential = pd.DataFrame({'Principal Component': highest_loadings.index,
                                 'Most Influential Feature': most_influential_features.values,
                                 'Loading Value': highest_loadings.values})

most_influential
```

```python
from sklearn.cluster import KMeans

# Fit and transform the feature data again since the state was reset
pca_features = pca.fit_transform(features)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(pca_features)
    wcss.append(kmeans.inertia_)

# Plotting the results onto a line graph to observe the 'elbow'
plt.figure(figsize=(20, 6))
plt.plot(range(1, 15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.grid(True)
plt.show()
```

```python
# Apply K-means clustering with k
kmeans = KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(pca_features)

# Visualize the clusters in the context of the first two principal components
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_features[:, 0], y=pca_features[:, 1], hue=cluster_labels, palette='viridis', s=100, alpha=0.7)
plt.title('Clusters of Car Models Based on PCA Features')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Assuming pca_features contains the PCA-transformed features and cluster_labels contains the cluster assignments

# Create a 3D scatter plot
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection='3d')

# Extract the first three principal components
x = pca_features[:, 0]
y = pca_features[:, 1]
z = pca_features[:, 2]

# Scatter plot using the first three principal components, colored by cluster label
scatter = ax.scatter(x, y, z, c=cluster_labels, cmap='viridis', s=50, alpha=0.7)

# Labeling axes
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D Visualization of Car Clusters')

# Legend
legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend1)

plt.show()
```

```python
import numpy as np

#========================================

random_car_code = 1000

#========================================

random_car_index = car_data.index[car_data['car_code'] == random_car_code][0]


random_car_cluster = cluster_labels[random_car_index]

# Identify other cars in the same cluster
same_cluster_indices = np.where(cluster_labels == random_car_cluster)[0]

# Calculate Euclidean distances from the random car to others in the same cluster
from scipy.spatial.distance import cdist
distances = cdist(pca_features[[random_car_index]], pca_features[same_cluster_indices], metric='euclidean').flatten()

# Find the indices of the ? closest cars (excluding the first one as it's the selected car itself)
closest_indices = same_cluster_indices[np.argsort(distances)[0:31]]

# Retrieve the car models for these closest cars
closest_car_models = car_data.iloc[closest_indices][['car_code', 'car_model']]
random_car_model = car_data.iloc[random_car_index]['car_model']

random_car_code, random_car_model, closest_car_models
```

```python
from sklearn.neighbors import NearestNeighbors


features_df = car_data.drop(columns=['car_code', 'car_model'])

knn = NearestNeighbors(n_neighbors=30)

knn.fit(features_df)

# Use a car's features as our query point
query = features_df.iloc[random_car_index].values.reshape(1, -1)

# Find the 5 nearest neighbors to the first car (excluding itself)
distances, indices = knn.kneighbors(query)

# Get the car codes of closest matches (including the query point itself)
closest_cars_indices = indices.flatten()
closest_cars = car_data.iloc[closest_cars_indices][['car_code', 'car_model']]

closest_cars
```

```python
price_df = pd.read_csv('car_prices_enriched_v2.csv')

car_code_list = closest_cars['car_code'].tolist()


# Filter data frames based on the list of car_code
df1_filtered = closest_cars[closest_cars['car_code'].isin(car_code_list)]
df2_filtered = price_df[price_df['car_code'].isin(car_code_list)]

# Merge the filtered data frames on car_code
merged_df = pd.merge(df1_filtered, df2_filtered, on='car_code', how='inner')

# Display the merged data frame
merged_df =  merged_df.drop_duplicates(subset=['car_manufacturer'])

print("KNN Results")
print(price_df.loc[price_df['car_code'] == random_car_code, ['car_code', 'car_manufacturer', 'car_model', 'car_last_price_in_brl']])
merged_df[['car_code', 'car_manufacturer', 'car_model_x', 'car_last_price_in_brl']]
```
